```python
import pandas as pd
df = pd.read_csv('cbb.csv')
df.head( )
```

```python
df.info()
```

```python
# set independent variable/attribute/predictor/feature sampai kolom AGE
X = df[df.columns[:23]]
 
# set dependent variable/clas/target/label
y = df['YEAR']
```

```python
# ini jadinya X

X
```

```python
# yang ini Y nya

y
```

```python
# terdapat fitur yang gak sama skalanya seperti ADJOE, G dan TEAM jadi perlu dibuat sama skalanya dengan StandardScaler()

from sklearn.preprocessing import StandardScaler
 
# standarisasi nilai-nilai dari dataset
scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)
```

MODELING

```python
# training 60% testing 40%

from sklearn.model_selection import train_test_split
 
# memisahkan data untuk training dan testing
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42)
```

```python
# fiting algoritma SVM dengan rbf kernal

from sklearn.svm import SVC
 
# membuat objek SVC dan memanggil fungsi fit untuk melatih model
clf = SVC(kernel='rbf')
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```
